### 3. compute DFT by convolution

```python
n = np.array(range(100))
x = np.cos(0.1*np.pi*n) + 0.5 * np.cos(np.pi*n)
h = np.ones(4) / 4
N = len(x) + len(h) - 1
x = np.concatenate((x, np.zeros(N - len(x))))
h = np.concatenate((h, np.zeros(N - len(h))))
y = np.fft.ifft(np.fft.fft(x) * np.fft.fft(h))
plt.figure(figsize=(18, 4))
plt.stem(y)
```

### plot frequency response

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
w1 = np.linspace(-np.pi, np.pi, 100)
w2 = np.linspace(-np.pi, np.pi, 100)
W1, W2 = np.meshgrid(w1, w2)
H = (1 + 2 * np.cos(W1) + 2 * np.cos(W2)) / 5
ax.plot_surface(W1, W2, H, rstride=1, cstride=1, cmap=plt.get_cmap('rainbow')) 
```
